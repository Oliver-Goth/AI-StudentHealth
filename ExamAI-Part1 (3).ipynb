## 6.7 Grouped Stress Levels 
For at forenkle klassifikationsopgaven grupperede vi de oprindelige 6 stressniveauer i 3 bredere kategorier:  
- **Low** (0 og 1)  
- **Medium** (2 og 3)  
- **High** (4 og 5)  

Denne tilgang reducerer overlap mellem klasserne og forbedrer fortolkeligheden, hvilket gør det muligt at fokusere på mere handlingsorienterede inden for *mental health groupings*.

```python
# 6.7.1 Create a new column for grouped stress levels
def group_stress_level(level):
    if level in [0, 1]:
        return 0  # Low
    elif level in [2, 3]:
        return 1  # Medium
    else:
        return 2  # High

df_clean['Stress_Grouped'] = df_clean['Stress_Level'].apply(group_stress_level)
```

```python
# 6.7.1.1 Check Class Distribution
df_clean['Stress_Grouped'].value_counts()
```

```python
# 6.7.1.2 Define Features and Grouped Target
X = df_clean.drop(columns=['Stress_Level', 'Stress_Grouped'])  # Features
y = df_clean['Stress_Grouped']  # Target = Grouped stress level
```

```python
# 6.7.1.3 Scale Features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 6.7.1.4 Split into Train and Test Sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

#### 6.7.1.x Random Forest (Grouped)

```python
# 6.7.1.5 Train Random Forest Model
from sklearn.ensemble import RandomForestClassifier

rf_model_grouped = RandomForestClassifier(random_state=42)
rf_model_grouped.fit(X_train, y_train)
```

```python
# 6.7.1.6 Make Predictions
y_pred_rf_grouped = rf_model_grouped.predict(X_test)
```

```python
# 6.7.1.7 Evaluate Performance
from sklearn.metrics import confusion_matrix, classification_report

print("Confusion Matrix (Random Forest – Grouped):")
print(confusion_matrix(y_test, y_pred_rf_grouped))

print("\nClassification Report (Random Forest – Grouped):")
print(classification_report(y_test, y_pred_rf_grouped))
```

##### 6.7.1.7.1 Resultater: Random Forest med grupperede stressniveauer  

Vi genoplærte Random Forest-klassifikatoren med de grupperede stressniveauer (Low, Medium, High). Modellen opnåede en **nøjagtighed på 40%**, en **markant forbedring** i forhold til de tidligere 22%, da vi brugte seks separate klasser.

Ydelsen var godt balanceret på tværs af de tre klasser. Grupperne med Lav (Low) og Mellem (Medium) stress havde lignende præcision og recall, mens **Høj stress** (High) viste den **højeste præcision (0.43)**, men en **lidt lavere recall (0.33)**.

Dette **bekræfter**, at en forenkling af **målvariablen har forbedret modellens præstation og tolkbarhed**.

## 7) Unsupervised Learning

### 7.1 Select features for clustering and Scale the data (required for K-Means)

```python
# 7.1.1 Prepare the Data
from sklearn.preprocessing import StandardScaler

# 7.1.2 Select relevant features (drop target labels)
unsupervised_features = df_clean.drop(columns=['Stress_Level', 'Stress_Grouped'])

# 7.1.3 Scale the features
scaler = StandardScaler()
X_unsup = scaler.fit_transform(unsupervised_features)
```

##### 7.1.3.1 Forbered data til Unsupervised Learning  
Vi fjernede alle target labels fra datasættet, så kun adfærds- og livsstilsrelaterede variabler var tilbage. Disse blev skaleret ved hjælp af `StandardScaler` for at forberede dataene til K-Means clustering, som er følsom over for forskelle i variablers skalaniveau.

### 7.2 Apply K-Means clustering

```python
from sklearn.cluster import KMeans

# 7.2.1 Choose number of clusters
k = 5  # 3 clusters to start

# 7.2.2 Fit K-Means
kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
clusters = kmeans.fit_predict(X_unsup)

# 7.2.3 Add clusters to your DataFrame
df_clean['Cluster_KMeans'] = clusters
```

Her anvender vi K-Means clustering til at opdele datasættet i 5 clusters baseret på skalerede features. Modellen initialiseres med n_init=10 for at sikre stabilitet og random_state=42 for reproducerbare resultater. Ved at bruge fit_predict() identificeres cluster labels, som gemmes i en ny kolonne ('Cluster_KMeans') i datasættet. Dette muliggør senere analyse og visualisering af de dannede grupper.

### 7.3 Visualize K-Means Clusters with PCA

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# 7.3.1 Reduce to 2D with PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_unsup)

# 7.3.2 Create a temporary DataFrame for plotting
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['Cluster'] = df_clean['Cluster_KMeans']

# 7.3.3 Plot with seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Cluster', palette='Set2', s=50)
plt.title("K-Means Clusters (Visualized with PCA)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

Denne kode anvender PCA til at reducere datasættet fra høj dimension til to komponenter, så clusterne kan visualiseres grafisk. PCA'en udtrækker de to mest betydningsfulde komponenter, som forklarer mest mulig varians i dataene, og gemmer dem som PC1 og PC2. Disse placeres i en ny DataFrame sammen med cluster labels fra K-Means-modellen.

Til sidst anvendes Seaborn til at tegne et scatterplot, hvor hver observation vises i det reducerede rum, farvekodet efter dens cluster. Denne visualisering gør det muligt at vurdere, hvordan K-Means har opdelt dataene, samt om clusterne er adskilte eller overlapper i det todimensionale rum.

### 7.3.4 Cluster Visualization (PCA)

```python
# Cluster Profiling: Interpret Cluster Centers
cluster_centers = kmeans.cluster_centers_
cluster_centers_df = pd.DataFrame(cluster_centers, columns=unsupervised_features.columns)
print("Cluster Centers (scaled features):")
print(cluster_centers_df)
```

```python
# Evaluate Clustering with Silhouette Score
from sklearn.metrics import silhouette_score
sil_score = silhouette_score(X_unsup, df_clean['Cluster_KMeans'])
print(f"Silhouette Score for K-Means clustering: {sil_score:.4f}")
```

Her henter vi cluster centers fra den trænede K-Means-model og konverterer dem til en DataFrame med samme kolonnenavne som de originale features. Disse centre repræsenterer de gennemsnitlige værdier for hver variabel i hver cluster og kan bruges til at fortolke, hvad der karakteriserer de enkelte cluster.

Herefter evalueres cluster kvaliteten med Silhouette Score, som måler, hvor klart adskilte clusterne er. Scoren går fra -1 til 1, hvor en høj værdi indikerer, at observationerne er godt matchet med deres egen cluster og dårligt matchet med nabo-cluster. Dette giver et kvantitativt mål for, hvor meningsfuld K-Means-inddelingen er.

En Silhouette Score på 0.0529 er relativt lav og indikerer derfor en dårlig cluster kvalitet, hvor clusterne ikke er klart adskilte, og observationerne ikke passer godt til deres tildelte cluster.

# 7.5 Cluster Profiling med Heatmap
